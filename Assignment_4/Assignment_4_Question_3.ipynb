```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
IC_data: pd.DataFrame = pd.read_csv("3_1_1_IC_Curve_Data.csv")
```

```python
colours = ["red", "blue", "green", "orange", "purple"]
linestyles = ["-", "--", "-.", ":", (0, (3, 1, 1, 1))]
order = [1, 2, 4, 6, 7]
for i in range(len(order)):
    plt.plot(IC_data[f"Gen{order[i]} MW"], IC_data[f"Gen{order[i]} $/MWh"], color=colours[i], label=f"Gen{order[i]}",
             linestyle=linestyles[i])

plt.minorticks_on()
plt.xlabel("Power Output (MW)")
plt.ylabel("Incremental Cost (Â£/MWh)")
plt.legend()
plt.grid(which="major", linestyle="--", linewidth="0.5")
plt.grid(which="minor", linestyle=":", linewidth="0.25")
plt.show()
```
