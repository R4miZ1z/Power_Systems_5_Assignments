```python
import pandas as pd
import matplotlib.pyplot as plt


# Read the data into a pandas DataFrame
df = pd.read_csv("data.csv")

df = df.fillna(0)

# Plot the data
plt.figure(figsize=(12, 6))
for column in df.columns[1:]:  # Skip the 'Year' column
    plt.plot(df['Year'], df[column] / 1000, marker='o', label=column)

plt.xlabel('Year')
plt.ylabel('Energy Output (TWh)')
plt.title('UK Energy Production by Source (2000-2023)')
plt.legend()
plt.minorticks_on()
plt.xticks(df['Year'], rotation=90)  # Ensure all years are displayed
plt.grid(which="major", linestyle='-', linewidth='0.5')
plt.grid(which="minor", linestyle=':', linewidth='0.25')
plt.savefig('IEA_UK_electricity_gen.pdf', format='pdf', bbox_inches='tight')
```

```python
# Define data
data = {
    "Electricity Source": ["Coal", "Oil", "Natural gas", "Biofuels", "Waste", "Nuclear", "Hydro", "Solar PV", "Wind", "Tidal"],
    "Value (GWh)": [4472, 2360, 96226, 29007, 9943, 40748, 7018, 13826, 81989, 11]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate total generation and percentage share
total_generation = df["Value (GWh)"].sum()
df["Percentage"] = (df["Value (GWh)"] / total_generation) * 100

df["Value (TWh)"] = (df["Value (GWh)"] / 1000)
print(df)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(df["Electricity Source"], df["Value (TWh)"], color='skyblue')
plt.xlabel("Electricity Source")
plt.ylabel("Generation (TWh)")
plt.title("Electricity Generation by Source in the UK (2023)")
plt.xticks(rotation=90, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.minorticks_on()
plt.grid(axis='y', which='both', linestyle=':', alpha=0.5)

# Add percentage labels
for i, (value, percentage) in enumerate(zip(df["Value (TWh)"], df["Percentage"])):
    plt.text(i, value + 1, f"{percentage:.3f}%", ha='center', fontsize=10)

# Show plot
plt.savefig('IEA_UK_electricity_gen_2023.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
